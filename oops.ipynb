```python
def scope_test():
    def do_local():
        spam = "local spam"

    def do_nonlocal():
        nonlocal spam
        spam = "nonlocal spam"

    def do_global():
        global spam
        spam = "global spam"

    spam = "test spam"
    do_local()
    print("After local assignment:", spam)
    do_nonlocal()
    print("After nonlocal assignment:", spam)
    do_global()
    print("After global assignment:", spam)

scope_test()
print("In global scope:", spam)
```

```python
class Mapping:
    def __init__(self, iterable):
        self.items_list = []
        self.__update(iterable)

    def update(self, iterable):
        for item in iterable:
            self.items_list.append(item)

    __update = update   # private copy of original update() method

class MappingSubclass(Mapping):

    def update(self, keys, values):
        # provides new signature for update()
        # but does not break __init__()
        for item in zip(keys, values):
            self.items_list.append(item)
m= Mapping([1, 2, 3])
print(m.items_list)
s = MappingSubclass([10, 20, 30])
print(s.items_list)

s.update(['key1', 'key2'], ['val1', 'val2'])
print(s.items_list)
```

```python
class Animal:
    def speak(self):
        raise NotImplementedError("Subclass must implement abstract method")

class Dog(Animal):
    def speak(self):
        return "Woof!"

class Cat(Animal):
    def speak(self):
        return "Meow!"

class Cow(Animal):
    def speak(self):
        return "Moo!"

# A generic function that uses Polymorphism
def animal_sound(animal):
    # This function doesn't need to know IF it's a dog or a cat.
    # It just calls .speak() and lets the object handle it.
    print(animal.speak())

# Creating objects
dog = Dog()
cat = Cat()
cow = Cow()

# Passing different forms (objects) to the same function
animal_sound(dog)  # Output: Woof!
animal_sound(cat)  # Output: Meow!
animal_sound(cow)  # Output: Moo!
```

```python
class Airplane:
    def fly(self):
        print("Engines on... Taking off!")

class Bird:
    def fly(self):
        print("Flapping wings... Soaring high!")

class Whale:
    def swim(self):
        print("Swimming deep.")

def lift_off(entity):
    # As long as the object has a .fly() method, this works.
    entity.fly()

p1 = Airplane()
p1.fly()
p2 = Bird()
p3 = Whale()

lift_off(p1) # Output: Engines on... Taking off!
lift_off(p2) # Output: Flapping wings... Soaring high!

# lift_off(p3) # This would CRASH because Whale has no .fly() method.
```

```python
class Parent:
    def speak(self):
        print("Parent is speaking generic words.")
        
class Child(Parent):
    # This is Method Overriding
    def speak(self):
        print("Child is shouting!")

p = Parent()
c = Child()

p.speak() # Output: Parent is speaking generic words.
c.speak() # Output: Child is shouting! (Parent's version is ignored)
```

```python
class Wallet:
    def __init__(self, money):
        self.money = money

    # This overloads the '+' operator
    def __add__(self, other_wallet):
        # We define that adding wallets returns a NEW wallet with the sum
        return Wallet(self.money + other_wallet.money)

    # This allows us to print the object nicely
    def __str__(self):
        return f"Wallet with ${self.money}"

# Creating objects
my_wallet = Wallet(100)
your_wallet = Wallet(50)

# The Magic Moment: Using '+' between objects
our_combined_wallet = my_wallet + your_wallet 

my_wallet.__str__

print(our_combined_wallet) 
# Output: Wallet with $150
```
