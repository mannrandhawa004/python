```python
import numpy as np

name = np.array(["mann", "suman", "leen"])

def cap(c):
    return c.upper()

# 1. Create the vectorized function (The Tool)
vectorized_func = np.vectorize(cap) 

# 2. USE the function on the array (The Action)
final_output = vectorized_func(name)  # <--- This is the missing step!

print(final_output)
# Output: ['MANN' 'SUMAN' 'LEEN']
```

```python
names = np.array(["arya", "ben", "cathy"])
def format_name(n):
    return "Student: " + n.capitalize()
vec_formatter = np.vectorize(format_name)
formatted_names = vec_formatter(names)

print(formatted_names)
```

```python
import numpy as np

codes = np.array(["A-101", "B-202", "C-303"])

def remove(ch):
    # "Replace '-' with nothing"
    return ch.replace("-", "") 

# Create the tool
vectorized_remove = np.vectorize(remove)

# Use the tool
formatted_string = vectorized_remove(codes)

print(formatted_string)
formatted_string
# Output: ['A101' 'B202' 'C303']
```

```python
scores = np.array([85, 40, 90, 30])

# Logic: Check scores > 50. If True -> "Pass", If False -> "Fail"
results = np.where(scores > 50, "Pass", "Fail")

print(results)
# Output: ['Pass' 'Fail' 'Pass' 'Fail']
```

```python
data = np.array([10, -1, 50, -1, 20])

# If data is -1, replace with 0. Otherwise, keep the original data (data)
clean_data = np.where(data == -1, 0, data)

print(clean_data)
# Output: [10  0 50  0 20]
```

```python
temps = np.array([20, 22, 19, 25, 30, 28, 24])
days = np.array(["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"])
high_temp = np.argmax(temps)
print(f"hightest temprature in the day",days[high_temp])
```

```python
sorted_indices = np.argsort(temps) #give the indexes
print(sorted_indices) 
ranked_days = days[sorted_indices]

print(ranked_days)
```

```python

arr = np.array([10, 20, 10, 30, 20, 50, 10])

# 1. Get values AND counts
values, counts = np.unique(arr, return_counts=True)
# values: [10, 20, 30, 50]
# counts: [ 3,  2,  1,  1]

# 2. Filter: Keep values where count is exactly 1
real_unique = values[counts == 1]

print(real_unique)
# Output: [30 50]
```
