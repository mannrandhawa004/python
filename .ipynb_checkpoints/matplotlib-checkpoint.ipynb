```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Prepare some data
x = np.linspace(0, 10, 200)
y_line = np.sin(x)
y_scatter = np.sin(x) + np.random.normal(0, 0.2, 200)
categories = ['Apples', 'Bananas', 'Cherries', 'Dates']
values = [15, 30, 45, 10]

# 2. Create the figure and subplots
# (2 rows, 2 columns)
fig, axs = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Matplotlib Essentials: Standard Plot Types', fontsize=20)

# --- Subplot 1: Line Plot ---
axs[0, 0].plot(x, y_line, color='blue', label='Sine Wave')
axs[0, 0].set_title('Line Plot (Trends)')
axs[0, 0].set_xlabel('Time (s)')
axs[0, 0].set_ylabel('Amplitude')
axs[0, 0].legend()
axs[0, 0].grid(True, linestyle='--', alpha=0.6)

# --- Subplot 2: Scatter Plot ---
axs[0, 1].scatter(x, y_scatter, color='red', alpha=0.5, label='Noisy Data')
axs[0, 1].set_title('Scatter Plot (Correlation)')
axs[0, 1].set_xlabel('Independent Variable')
axs[0, 1].set_ylabel('Measured Value')
axs[0, 1].legend()

# --- Subplot 3: Bar Chart ---
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
axs[1, 0].bar(categories, values, color=colors, edgecolor='black')
axs[1, 0].set_title('Bar Chart (Comparisons)')
axs[1, 0].set_xlabel('Fruit Type')
axs[1, 0].set_ylabel('Quantity Sold')

# --- Subplot 4: Histogram ---
data_dist = np.random.randn(1000)
axs[1, 1].hist(data_dist, bins=30, color='purple', alpha=0.7, edgecolor='white')
axs[1, 1].set_title('Histogram (Distribution)')
axs[1, 1].set_xlabel('Value Range')
axs[1, 1].set_ylabel('Frequency')

# 3. Adjust layout so titles don't overlap
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

X, Y = np.meshgrid(np.linspace(-5, 5, 50), np.linspace(-5, 5, 50))
Z = np.sqrt(X**2 + Y**2)

surf = ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title('3D Surface Plot')
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z (Depth)')
fig.colorbar(surf, shrink=0.5, aspect=5) # Adds a color scale bar
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Laptops_Sold': [45, 52, 48, 60, 55, 72, 80, 75, 68, 155, 120, 150],
    'Phones_Sold': [120, 110, 130, 145, 140, 160, 155, 150, 140, 165, 210, 250],
    'Ad_Spend': [1000, 1200, 1100, 1500, 1400, 1800, 2000, 1900, 1700, 2200, 3000, 4000],
    'Region': ['East', 'West', 'East', 'West', 'East', 'West', 'East', 'West', 'East', 'West', 'East', 'West']
}

df = pd.DataFrame(data)
```

```python
months = df['Month']
laptop_sold = df['Laptops_Sold']

plt.plot(months,laptop_sold,color="green",label="Laptop Sales",linestyle="--")
plt.title("Laptop Sales Trends")
plt.grid(True)
plt.xlabel("Months")
plt.ylabel("No.of Laptops sold")
plt.ylim(40,200)
plt.legend()
plt.show()
```

```python
months = df['Month']
laptop_sold = df['Laptops_Sold']
phone_sold = df['Phones_Sold']

plt.plot(months, laptop_sold, linestyle="--", label="Laptop")
plt.plot(months, phone_sold, linestyle="--", label="Phones")

plt.grid(True)
plt.xlabel("Months")
plt.ylabel("Sales")
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 5))

ax.plot(df["Month"], df["Laptops_Sold"], label="Laptops", linestyle="--")
ax.plot(df["Month"], df["Phones_Sold"], label="Phones", linestyle="-.")

ax.set_xlabel("Months")
ax.set_ylabel("Sales")
ax.legend()
ax.grid(True)

plt.show()
```

```python
plt.bar(df["Month"],df["Ad_Spend"],color="green",zorder=3)
plt.xlabel("Months",fontweight='bold')
plt.grid(True)
plt.ylabel("Spend of each month",fontweight='bold')
plt.show()
```

```python
plt.scatter(df["Ad_Spend"],df["Phones_Sold"],marker="*")
plt.xlabel("Spend",fontweight='bold')
plt.grid(True)
plt.ylabel("Phone sales",fontweight='bold')
plt.show()
```

```python
plt.scatter(
    df["Ad_Spend"],
    df["Phones_Sold"],
    label="Phone Sales",
    marker="*",
    color="blue"
)

plt.xlabel("Ad Spend", fontweight="bold")
plt.ylabel("Phone Sales", fontweight="bold")
plt.legend()
plt.grid(True,linestyle='--', alpha=0.6)
plt.show()
```

```python
plt.hist(df["Laptops_Sold"])
plt.title("Histogram")
plt.xlabel("Laptops Sold")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()

```

```python
plt.plot(
    df["Month"],
    df["Laptops_Sold"],
    marker="o",
    label="Laptops Sold"
)

plt.plot(
    df["Month"],
    df["Phones_Sold"],
    linestyle="--",
    marker="s",
    label="Phones Sold"
)

# Labels & grid
plt.xlabel("Months", fontweight="bold")
plt.ylabel("Sales", fontweight="bold")
plt.title("Laptop vs Phone Sales Comparison")
plt.grid(True,linestyle="--")
plt.legend()

plt.show()
```

```python
df["Total_Sales"] = df["Laptops_Sold"] + df["Phones_Sold"]
region_sales = df.groupby("Region")["Total_Sales"].sum()

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# -------- Left: Pie Chart --------
axes[0].pie(
    region_sales,
    labels=region_sales.index,
    autopct="%1.1f%%",
    startangle=90
)
axes[0].set_title("Total Sales by Region")

# -------- Right: Bar Chart --------
axes[1].bar(
    df["Month"],
    df["Ad_Spend"],
    zorder=3
)
axes[1].set_title("Monthly Ad Spend")
axes[1].set_xlabel("Months",fontweight="bold")
axes[1].set_ylabel("Ad Spend",fontweight="bold")
axes[1].grid(True, axis="y")

# Layout adjustment
plt.tight_layout()
plt.show()
```

```python

df["Total_Sales"] = df["Laptops_Sold"] + df["Phones_Sold"]

# Group data for pie chart
region_sales = df.groupby("Region")["Total_Sales"].sum()

# ------------------- Create Subplots -------------------
fig, axes = plt.subplots(3, 2, figsize=(16, 16))  # 3 rows, 2 columns
fig.suptitle("Your Main Figure Title", fontsize=16, fontweight='bold')


# 1️⃣ Line Plot (Laptop & Phone sales)
axes[0, 0].plot(df["Month"], df["Laptops_Sold"], marker='o', label="Laptops")
axes[0, 0].plot(df["Month"], df["Phones_Sold"], linestyle="--", marker='s', label="Phones")
axes[0, 0].set_title("Laptop vs Phone Sales")
axes[0, 0].set_ylabel("Units Sold")
axes[0, 0].legend()
axes[0, 0].grid(True)

# 2️⃣ Bar Chart (Ad Spend)
axes[0, 1].bar(df["Month"], df["Ad_Spend"], color='green',zorder=3)
axes[0, 1].set_title("Monthly Ad Spend")
axes[0, 1].set_ylabel("Ad Spend")
axes[0, 1].grid(True)

# 3️⃣ Scatter Plot (Ad Spend vs Phone Sales)
axes[1, 0].scatter(df["Ad_Spend"], df["Phones_Sold"], color='orange', marker='^')
axes[1, 0].set_title("Ad Spend vs Phone Sales")
axes[1, 0].set_xlabel("Ad Spend")
axes[1, 0].set_ylabel("Phones Sold")
axes[1, 0].grid(True)

# 4️⃣ Histogram (Laptop Sales)
axes[1, 1].hist(df["Laptops_Sold"], bins=6, edgecolor="black", color='purple',zorder=3)
axes[1, 1].set_title("Laptop Sales Distribution")
axes[1, 1].set_xlabel("Laptops Sold")
axes[1, 1].set_ylabel("Frequency")
axes[1, 1].grid(axis="y")

# 5️⃣ Pie Chart (Total Sales by Region)
axes[2, 0].pie(region_sales, labels=region_sales.index, autopct="%1.1f%%", startangle=90)
axes[2, 0].set_title("Total Sales by Region")

# 6️⃣ Box Plot (Laptop Sales)
axes[2, 1].boxplot(df["Laptops_Sold"], patch_artist=True, boxprops=dict(facecolor='lightblue'))
axes[2, 1].set_title("Laptop Sales Boxplot")
axes[2, 1].set_ylabel("Units Sold")
axes[2, 1].grid(axis="y")

# Layout adjustment
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
x = range(1, 13)
y1 = [45,52,48,60,55,72,80,75,68,85,120,150]
y2 = [120,110,130,145,140,160,155,150,140,165,210,250]
y3 = [1000,1200,1100,1500,1400,1800,2000,1900,1700,2200,3000,4000]
y4 = [5,7,6,8,7,9,10,8,9,11,12,14]
y5 = [20,25,22,28,26,30,32,29,27,33,35,40]
y6 = [50,55,52,60,58,65,70,66,64,72,80,90]

fig, axes = plt.subplots(3, 3, figsize=(18, 12))

axes[0,0].plot(x, y1, marker='o'); axes[0,0].set_title("Plot 1")
axes[0,1].plot(x, y2, marker='s'); axes[0,1].set_title("Plot 2")
axes[0,2].bar(x, y3, color='green'); axes[0,2].set_title("Plot 3")

axes[1,0].scatter(x, y4, color='orange'); axes[1,0].set_title("Plot 4")
axes[1,1].hist(y5, bins=5, color='purple'); axes[1,1].set_title("Plot 5")
axes[1,2].boxplot(y6, patch_artist=True); axes[1,2].set_title("Plot 6")

axes[2,1].axis('off')
axes[2,2].axis('off')
axes[2,0].axis('off')

plt.tight_layout()
plt.show()
```
