```python
data = "  apple, banana, cherry  "
clean_data = data.strip()             # Removes surrounding spaces
fruits = clean_data.split(", ")       # Splits into a list

print(fruits)                         # Output: ['apple', 'banana', 'cherry']
print(" - ".join(fruits))             # Output: "apple - banana - cherry"
```
