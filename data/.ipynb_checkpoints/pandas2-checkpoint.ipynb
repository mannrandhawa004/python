```python
# errors='coerce' turns the impossible date (2025-15-40) into NaT (Not a Time/Null)
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
```

```python
# Fill NaT with the most common date (mode)
# most_common_date = df['Date'].mode()[0]
# df['Date'] = df['Date'].fillna(most_common_date)

# print("NaT values filled with:", most_common_date)

# Convert to string and fill NaT with a custom message
df['Date'] = df['Date'].dt.strftime('%Y-%m-%d').fillna("Invalid/Missing Date")

print(df['Date'].value_counts())
```

```python
df.head()
```

```python
print(df[['Price', 'Quantity']].describe())
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Forward fill: fill NaT with the value from the previous row
df['Date'] = df['Date'].ffill()

print(f"Remaining NaT: {df['Date'].isna().sum()}")
```

```python
# Option B: Replace negative prices with the median (better for keeping data)
median_price = df[df['Price'] > 0]['Price'].median()
df.loc[df['Price'] <= 0, 'Price'] = median_price
```

```python
import matplotlib.pyplot as plt

# Create a boxplot to see the "dots" (outliers)
plt.boxplot(df['Price'].dropna())
plt.title("Price Boxplot (Before Cleaning)")
plt.show()
```

```python
df["Category"] = df["Category"].str.strip().str.capitalize()
```

```python
df.head()
```

```python
# 1. First, strip any hidden spaces and make it consistent case (optional but recommended)
df['Category'] = df['Category'].str.strip()

# 2. Define a dictionary of { "Old Value": "New Value" }
mapping = {
    'Elec.': 'Electronics'
}

df['Category'] = df['Category'].replace(mapping)

# 4. Check the results
print(df['Category'].unique())
```

```python
df = df.dropna(subset=['Email'])
```

```python
# Find the most common value (usually False)
discount_mode = df['Discount_Applied'].mode()[0]
discount_mode
```

```python
df['Discount_Applied'] = df['Discount_Applied'].fillna(discount_mode)
```

```python
print("Final Null Count:")
print(df[['Email', 'Discount_Applied']].isnull().sum())
```

```python
# Identify the threshold (e.g., anything over 20 is suspicious for this store)
quantity_median = df['Quantity'].median()

# Replace suspicious quantities with the median
df.loc[df['Quantity'] > 20, 'Quantity'] = quantity_median

# Verify again
print(df['Quantity'].describe())
```
