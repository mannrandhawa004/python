```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Prepare some data
x = np.linspace(0, 10, 200)
y_line = np.sin(x)
y_scatter = np.sin(x) + np.random.normal(0, 0.2, 200)
categories = ['Apples', 'Bananas', 'Cherries', 'Dates']
values = [15, 30, 45, 10]

# 2. Create the figure and subplots
# (2 rows, 2 columns)
fig, axs = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Matplotlib Essentials: Standard Plot Types', fontsize=20)

# --- Subplot 1: Line Plot ---
axs[0, 0].plot(x, y_line, color='blue', label='Sine Wave')
axs[0, 0].set_title('Line Plot (Trends)')
axs[0, 0].set_xlabel('Time (s)')
axs[0, 0].set_ylabel('Amplitude')
axs[0, 0].legend()
axs[0, 0].grid(True, linestyle='--', alpha=0.6)

# --- Subplot 2: Scatter Plot ---
axs[0, 1].scatter(x, y_scatter, color='red', alpha=0.5, label='Noisy Data')
axs[0, 1].set_title('Scatter Plot (Correlation)')
axs[0, 1].set_xlabel('Independent Variable')
axs[0, 1].set_ylabel('Measured Value')
axs[0, 1].legend()

# --- Subplot 3: Bar Chart ---
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
axs[1, 0].bar(categories, values, color=colors, edgecolor='black')
axs[1, 0].set_title('Bar Chart (Comparisons)')
axs[1, 0].set_xlabel('Fruit Type')
axs[1, 0].set_ylabel('Quantity Sold')

# --- Subplot 4: Histogram ---
data_dist = np.random.randn(1000)
axs[1, 1].hist(data_dist, bins=30, color='purple', alpha=0.7, edgecolor='white')
axs[1, 1].set_title('Histogram (Distribution)')
axs[1, 1].set_xlabel('Value Range')
axs[1, 1].set_ylabel('Frequency')

# 3. Adjust layout so titles don't overlap
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

X, Y = np.meshgrid(np.linspace(-5, 5, 50), np.linspace(-5, 5, 50))
Z = np.sqrt(X**2 + Y**2)

surf = ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title('3D Surface Plot')
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z (Depth)')
fig.colorbar(surf, shrink=0.5, aspect=5) # Adds a color scale bar
plt.show()
```

```python
days = [1, 2, 3, 4, 5]
organic = [2, 3, 4, 3, 2]
ads = [1, 1, 1, 2, 2]

plt.stackplot(days, organic, ads, labels=['Organic', 'Ads'], colors=['#8bc34a', '#2196f3'])
plt.legend(loc='upper left')
plt.title("Traffic Sources Over 5 Days")
```
