```python
def evenlyDivides(n):
        count = 0
        # Convert the number to a string so we can loop through each digit
        temp_str = str(n)
        
        for digit_char in temp_str:
            digit = int(digit_char) # Convert the character back to an integer
            
            # 1. Check if the digit is not zero (to avoid division by zero)
            # 2. Check if n is evenly divisible by the digit
            if digit != 0 and n % digit == 0:
                count += 1 # Python way to increment
                
        print(count)
evenlyDivides(2446)
```

```python
def half_pyramid_pattern(n):
    for i in range(1,n+1):
        for j in range(1,i+1):
            print("*",end="")
        print()
half_pyramid_pattern(5)
```
