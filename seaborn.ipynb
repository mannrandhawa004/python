```python
import seaborn as sns
sns.set_theme(style="darkgrid")
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
sns.relplot(data=tips,x="total_bill",y="tip",kind="scatter",palette="plasma",hue="sex")
```

```python
# Create a visualization
sns.relplot(
    data=tips,
    x="total_bill", y="tip", col="time",
    hue="smoker", style="smoker", size="size",
)
```

```python
dots = sns.load_dataset("dots")
sns.relplot(data=dots,kind="line",x="time", y="firing_rate", col="align",
    hue="choice", size="coherence", style="choice",
    facet_kws=dict(sharex=False)
)
```

```python
fmri = sns.load_dataset("fmri")
sns.relplot(
    data=fmri, kind="line",
    x="timepoint", y="signal", col="region",
    hue="event", style="event",
)
```

```python
sns.lmplot(data=tips, x="total_bill", y="tip", col="time", hue="sex")
```

```python
sns.displot(data=tips, x="total_bill", col="time", kde=True)
```

```python
sns.displot(data=tips, x="total_bill", col="time", hue="smoker",kde=True,multiple="stack")
```

```python
sns.catplot(data=tips, kind="swarm", x="day", y="total_bill", hue="smoker")
```

```python
sns.catplot(data=tips, kind="violin", x="day", y="total_bill")
```

```python
sns.catplot(data=tips, kind="bar", x="day", y="total_bill", hue="smoker")
```

```python
flights = sns.load_dataset("flights")
flights
```

```python
sns.relplot(data=flights, x="year", y="passengers", hue="month", kind="line")
```

```python
sns.barplot(data=flights, x="year", y="passengers")
```

```python
flights_wide = flights.pivot(index="year", columns="month", values="passengers")
flights_wide.head()
```

```python
sns.relplot(data=flights_wide)
```

```python
sns.relplot(data=flights, x="month", y="passengers", hue="year", kind="line")
```

```python
sns.catplot(data=flights_wide, kind="box")
```
