```python
#renaming the column name
details_dataset.rename(columns={"Order ID" :"Order_Id"}, inplace=True)
details_dataset
```

```python
details_dataset.isnull().head()
details_dataset.dropna(axis=0) # drop a row that contains any null value 
# details_dataset.drop(details_dataset.iloc[0])
```

```python
details_dataset["Category"] = details_dataset["Category"].apply(lambda x: x.upper())
details_dataset
```

```python
#creating a total_amount column
details_dataset["Total"] = details_dataset["Amount"] * details_dataset["Quantity"]
details_dataset.head()
```

Aggregation and Grouping

```python
details_dataset.groupby("Category")["Profit"].mean()
```

```python
# Simple average of the whole column
avg_profit = details_dataset["Profit"].mean()
print(avg_profit)
```

```python
#COUNT THE NUMBER OF OCCURENCE
payment_counts = details_dataset["PaymentMode"].value_counts()
payment_counts
```

```python
# This creates a matrix of Categories vs PaymentModes
pd.crosstab(details_dataset['Category'], details_dataset['PaymentMode'])
```
